# Convert the Keras model to a TensorFlow Lite model and write the .tflite file

```
keras_model_path = r'drive/My Drive/Colab Notebooks/ngocnet.h5'
tfjs_model_path = r'drive/My Drive/Colab Notebooks/ngocnet.tfjs'
```

```
import numpy as np
model = tf.keras.models.load_model(keras_model_path)
tfjs.converters.save_keras_model(model, tfjs_model_path)
```
